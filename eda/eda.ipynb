```python
import pandas as pd

# 1) Ma'lumotni yuklash
df = pd.read_csv("C:/Users/zafki/Desktop/futbol_pozitsiya_modeli/data/players.csv")

# 2) Birinchi 5 qatorni ko'rish
display(df.head())

# 3) Ustunlar va turlarini tekshirish
print(df.info())

# 4) Raqamli ustunlar bo‘yicha statistik xulosa
display(df.describe())
```

```python
df.isna().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Faqat raqamli ustunlar
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f")
plt.title("Raqamli ustunlar orasidagi korrelyatsiya")
plt.show()


```

```python
sns.countplot(data=df, x='Position')

plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Position')  # <-- bu yerda 'Position' to‘g‘ri
plt.title("O'yinchi pozitsiyalari bo'yicha taqsimot")
plt.xlabel("Pozitsiya")
plt.ylabel("O'yinchilar soni")
plt.show()

print(df.columns)
```

```python
print(df.columns.tolist())


# Ustun nomlarini tozalash
df.columns = df.columns.str.strip()

# Tozalangan ustunlarni tekshirish
print(df.columns.tolist())  # bu yerda 'Overall' bo'lishi kerak

# O'rtacha Overall hisoblash
avg_overall_by_pos = df.groupby('Position')['Overall'].mean().sort_values(ascending=False)
print(avg_overall_by_pos)

# Vizualizatsiya
plt.figure(figsize=(10,6))
sns.barplot(x=avg_overall_by_pos.index, y=avg_overall_by_pos.values, palette="viridis")
plt.title("Har bir pozitsiya bo'yicha o'rtacha Overall")
plt.xlabel("Pozitsiya")
plt.ylabel("O'rtacha Overall")
plt.xticks(rotation=45)
plt.show()

# Eng kuchli o'yinchilar
best_players_by_pos = df.loc[df.groupby('Position')['Overall'].idxmax()][['Name', 'Position', 'Overall']]
print(best_players_by_pos)
```
